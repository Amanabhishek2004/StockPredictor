```python
# changing string cols to actual list
import numpy as np
def process_row(row):
        specific_column_data = row.replace("[" , "").replace("]" , "")
        data = specific_column_data.split(",") 

        return [float(x)  if x !="nan" else np.nan for x in data][:-1]  if len(data) > 4 else [float(x)  if x !="nan" else np.nan for x in data]
```

```python

def CalculateCAGR(array) :
      #  start and end both false then return -1 
      start = array[0]
      # print(len(array))
      end = array[-1]
      if (start>0 and end>0)  or (start < 0  and end < 0) :
         return  (start/end)**(1/len(array))-1
      # print(( (start/end)**(1/len(array)))-1)
      if start < 0  :
         return 0.0
      if end < 0 :
         return 1.0
      return -1.0 
```

```python
def CalculateAverageGrowth(array) :
    #  no need to check for nan for np,diff() since it takes care  
    diffrence = -1*np.diff(array)
    # Check for NaN values or zeros
    if np.isnan(diffrence).any() or (diffrence == 0).all():
        return -1
    
    for index , i  in enumerate(diffrence) :
        if array[index + 1] == 0:
            diffrence[index] = -1 # Avoid division by zero
        else :
            diffrence[index] /= array[index+1]
    
    # Check if the resulting array is empty or contains NaN values
    if diffrence.size == 0 or np.isnan(diffrence).any():
        return -1
    return np.mean(diffrence)
                
```

```python
import numpy as np
for columns in stringed_columns:
    convertedstringedcolumns = companyexcelsheet[columns].apply(process_row)
    companyexcelsheet[f"{columns}_converted"] = convertedstringedcolumns
    cagrcalculation = companyexcelsheet[f"{columns}_converted"].apply( CalculateCAGR )
    companyexcelsheet[f"{columns}_cagr"] = cagrcalculation
    avggrowthcalculation = companyexcelsheet[f"{columns}_converted"].apply(CalculateAverageGrowth)
    companyexcelsheet[f"{columns}_avggrowth"] = avggrowthcalculation
    
```
